```python
import pandas as pd
pd.set_option('display.max_columns', 100)
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_csv('job opportunities.csv')
df.head(12)
```

```python
df.info()
```

```python
df = df.replace("NaN",0)
```

```python
df = df.fillna(0)
```

```python
df.head(12)
```

```python
df["SCALAR_FACTOR"].unique()
```

```python
#droping the useless coloumns
df = df.drop(['SCALAR_FACTOR', 'SCALAR_ID','SYMBOL','TERMINATED'], axis=1)
```

```python
df['Statistics'].unique()
```

```python
df['Job vacancies'] = df.loc[df['Statistics'] == 'Job vacancies']['VALUE']
df.head()
```

```python
df['Payroll employees'] = df.loc[df['Statistics'] == 'Payroll employees']['VALUE']
df.head()
```

```python
df['Job vacancy rate'] = df.loc[df['Statistics'] == 'Job vacancy rate']['VALUE']
df.head()
```

```python
df['Payroll employees'] = df['Payroll employees'].shift(-1)
```

```python
df['Job vacancy rate'] = df['Job vacancy rate'].shift(-2)
```

```python
df.head(12)
```

```python
df=df.drop(['Statistics', 'VALUE'], axis=1)
```

```python
df = df.dropna()
```
