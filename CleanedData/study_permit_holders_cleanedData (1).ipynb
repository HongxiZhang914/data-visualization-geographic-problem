```python
import pandas as pd
pd.set_option('display.max_columns', 100)
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
df = pd.read_excel('study permit holders.xlsx')
df.head(12)
```

```python
df.info()
```

```python
for i in range(1,156):
    month = ["Jan","Feb","Mar","Q1Total","Apr","May","Jun","Q2Total","July","Aug","Sep","Q3Total","Oct","Nov","Dec","Q4total","YearTotal"]
    df[str(2015+int((i-1)/17))+month[(i+16)%17]] = df["Unnamed: "+str(i)]
```

```python
df.head()
```

```python
df = df.drop([0,1,2,3])
df.head()
```

```python
df = df.drop(df.iloc[:,1:156],axis = 1)
df.head()
```

```python
df = df.drop(columns = ["Unnamed: 156"])
df.head()
```

```python
df.tail(12)
```

```python
df = df.drop([221,222,223,224,225])
df.tail(12)
```

```python
df = df.drop([226,227,228])
df.tail(12)
```

```python
df = df.replace("--",0)
df.tail(12)
```

```python
df.dtypes
```

```python
df = df.astype(str)
```

```python
df.dtypes
```

```python
#change data type to int
for column in df:
    try:
        df[column] = df[column].str.replace(",","").astype(int)
    except:
        pass
```

```python
df = df.drop(df[(df["2015YearTotal"] <= 0) & (df["2023YearTotal"]<=0)].index)
df.head(12)
```
