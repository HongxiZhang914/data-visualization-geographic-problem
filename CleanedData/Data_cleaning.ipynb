```python
import pandas as pd
pd.set_option('display.max_columns', 100)
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns

```

```python
df = pd.read_csv('Canadian_population_distribution.csv')
df.head(12)
```

```python
#delete columns with no values
df = df.drop(columns = ['UOM','VECTOR','STATUS','SYMBOL','TERMINATED'])
df.info()
```

```python
#check if the column has only 0 value in all rows
df['UOM_ID'].unique()
```

```python
df['SCALAR_ID'].unique()
```

```python
#drop these columns since they have no data insides these columns
df = df.drop(columns = ['UOM_ID','SCALAR_ID'])
df.info()
```

```python
#check duplicate values
duplicate_value = df[df.duplicated()]
duplicate_value
```

```python
#data type looks perfect at this stage, no need to change
```

```python
#I found the table has a strange column called Population and dwelling counts (11), so I choose to check the values inside
df['Population and dwelling counts (11)'].unique()

```

```python
#to understanding this table, it is better to transfer these 11 kind of values into new columns with same names
df['Population, 2021'] = df.loc[df['Population and dwelling counts (11)'] == 'Population, 2021']['VALUE']
df.head()
```

```python
#Continue with other 10 columns
df['Population, 2016'] = df.loc[df['Population and dwelling counts (11)'] == 'Population, 2016']['VALUE']
df.head()
```

```python
df['Population percentage change, 2016 to 2021'] = df.loc[df['Population and dwelling counts (11)'] == 'Population percentage change, 2016 to 2021']['VALUE']
df.head()
```

```python
df['Total private dwellings, 2021'] = df.loc[df['Population and dwelling counts (11)'] == 'Total private dwellings, 2021']['VALUE']
df.head()
```

```python
df['Total private dwellings, 2016'] = df.loc[df['Population and dwelling counts (11)'] == 'Total private dwellings, 2016']['VALUE']
df.head()
```

```python
df['Total private dwellings percentage change, 2016 to 2021'] = df.loc[df['Population and dwelling counts (11)'] == 'Total private dwellings percentage change, 2016 to 2021']['VALUE']
df.head(6)
```

```python
df['Private dwellings occupied by usual residents, 2021'] = df.loc[df['Population and dwelling counts (11)'] == 'Private dwellings occupied by usual residents, 2021']['VALUE']
df.tail(6)
```

```python
df['Private dwellings occupied by usual residents, 2016'] = df.loc[df['Population and dwelling counts (11)'] == 'Private dwellings occupied by usual residents, 2016']['VALUE']
df.tail()
```

```python
df['Private dwellings occupied by usual residents percentage change, 2016 to 2021'] = df.loc[df['Population and dwelling counts (11)'] == 'Private dwellings occupied by usual residents percentage change, 2016 to 2021']['VALUE']
df.tail()
```

```python
df['Land area in square kilometres, 2021'] = df.loc[df['Population and dwelling counts (11)'] == 'Land area in square kilometres, 2021']['VALUE']
df.tail()
```

```python
df['Population density per square kilometre, 2021'] = df.loc[df['Population and dwelling counts (11)'] == 'Population density per square kilometre, 2021']['VALUE']
df.tail()
```

```python
#As I already transfer all the data in Population and dwelling counts (11) column and value column into new columns, it is time to delete old columns
df = df.drop(columns = ['Population and dwelling counts (11)', 'VALUE'])
df.info()
        
```

```python
#Now we can see there are only 14 rows for each new column, that is because I divided 154 values into 11 columns
#But I did not lose values because the total number of values are still 154, to make the table readable,
#I need to delete duplicate rows now, ex. the 'Canada' value appear 11 times in GEO column
#first I need to do is to make sure all the non-null values appear in the same row.
#Shift method is a good option
df['Population, 2016'] = df['Population, 2016'].shift(-1)

```

```python
df.head(12)
```

```python
df['Population percentage change, 2016 to 2021'] = df['Population percentage change, 2016 to 2021'].shift(-2)
```

```python
df['Total private dwellings, 2021'] = df['Total private dwellings, 2021'].shift(-3)
```

```python
df['Total private dwellings, 2016'] = df['Total private dwellings, 2016'].shift(-4)
```

```python
df['Total private dwellings percentage change, 2016 to 2021'] = df['Total private dwellings percentage change, 2016 to 2021'].shift(-5)
```

```python
df['Private dwellings occupied by usual residents, 2021'] = df['Private dwellings occupied by usual residents, 2021'].shift(-6)
```

```python
df['Private dwellings occupied by usual residents, 2016'] = df['Private dwellings occupied by usual residents, 2016'].shift(-7)
```

```python
df['Private dwellings occupied by usual residents percentage change, 2016 to 2021'] = df['Private dwellings occupied by usual residents percentage change, 2016 to 2021'].shift(-8)
```

```python
df['Land area in square kilometres, 2021'] = df['Land area in square kilometres, 2021'].shift(-9)
```

```python
df['Population density per square kilometre, 2021'] = df['Population density per square kilometre, 2021'].shift(-10)
```

```python
df.head(12)
```

```python
#Now the table is readable, now we need to delete those rows with null values, since all the values all transfer to the same row.
df = df.dropna()
```

```python
df.head(14)
```

```python
#Now the table is more readable and organized. We do not need to much rows, but we need the table to be clear to look at.
#the last thing need to do is delete REF_DATE column, since we got REF_DATE in other column names.
df = df.drop(columns = ['REF_DATE'])
```
